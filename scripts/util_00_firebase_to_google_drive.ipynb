```python
import os
import firebase_admin
from firebase_admin import credentials, storage
from google.colab import drive

drive.mount('/content/drive')
```

```python
cred_path = '/content/drive/MyDrive/Dataset/serviceAccountKey.json'
bucket_name = 'fairplayfairy-3e2eb.firebasestorage.app'

try:
    # Firebase ì•±ì´ ì´ë¯¸ ì´ˆê¸°í™”ë˜ì—ˆëŠ”ì§€ í™•ì¸í•˜ê³ , ê·¸ë ‡ì§€ ì•Šì€ ê²½ìš°ì—ë§Œ ì´ˆê¸°í™”
    if not firebase_admin._apps:
        cred = credentials.Certificate(cred_path)
        firebase_admin.initialize_app(cred, {
            'storageBucket': bucket_name
        })
    print("âœ… Firebase ì¸ì¦ì— ì„±ê³µí–ˆìŠµë‹ˆë‹¤.")
except Exception as e:
    print(f"ðŸ”¥ Firebase ì¸ì¦ ì˜¤ë¥˜: {e}")

bucket = storage.bucket()
print(f"âœ… í˜„ìž¬ APIê°€ í˜¸ì¶œí•˜ëŠ” ë²„í‚· ì´ë¦„: {bucket.name}")
```

```python
# --- ì„¤ì • ê°’ ---
# Firebase Storageì—ì„œ íŒŒì¼ì„ ê°€ì ¸ì˜¬ í´ë” ê²½ë¡œ (ì „ì²´ íŒŒì¼ì„ ì›í•˜ë©´ ''ë¡œ ë¹„ì›Œë‘ì„¸ìš”)
FIREBASE_FOLDER_PATH = 'videos/'

# íŒŒì¼ì„ ì €ìž¥í•  Google Drive ê²½ë¡œ
GDRIVE_SAVE_PATH = '/content/drive/MyDrive/Dataset/videos'
# ----------------

# Google Driveì— ì €ìž¥í•  í´ë” ìƒì„±
os.makedirs(GDRIVE_SAVE_PATH, exist_ok=True)

try:
    # Firebase Storageì—ì„œ íŒŒì¼ ëª©ë¡ ê°€ì ¸ì˜¤ê¸°
    blobs = bucket.list_blobs(prefix=FIREBASE_FOLDER_PATH)
    print(f"ì´ {len(list(blobs))}ê°œì˜ íŒŒì¼ì„ í™•ì¸í–ˆìŠµë‹ˆë‹¤. ë‹¤ìš´ë¡œë“œë¥¼ ì‹œìž‘í•©ë‹ˆë‹¤...")

    # ë‹¤ì‹œ ëª©ë¡ì„ ê°€ì ¸ì™€ ìˆœíšŒ
    blobs = bucket.list_blobs(prefix=FIREBASE_FOLDER_PATH)

    for blob in blobs:
        # í´ë” ìžì²´ëŠ” ê±´ë„ˆë›°ê¸°
        if blob.name.endswith('/'):
            continue

        # ë‹¤ìš´ë¡œë“œí•  íŒŒì¼ ê²½ë¡œ ì„¤ì •
        destination_file_name = os.path.join(GDRIVE_SAVE_PATH, os.path.basename(blob.name))

        try:
            # íŒŒì¼ ë‹¤ìš´ë¡œë“œ
            blob.download_to_filename(destination_file_name)
            print(f"ðŸ“¥ '{blob.name}' -> '{destination_file_name}'ë¡œ ë‹¤ìš´ë¡œë“œ ì™„ë£Œ")
        except Exception as e:
            print(f"â—ï¸ '{blob.name}' ë‹¤ìš´ë¡œë“œ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")

    print("\nðŸŽ‰ ëª¨ë“  íŒŒì¼ ì „ì†¡ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤!")

except Exception as e:
    print(f"ðŸ”¥ íŒŒì¼ ëª©ë¡ì„ ê°€ì ¸ì˜¤ëŠ” ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}")
```
